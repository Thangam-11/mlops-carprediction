```python
# check the data types of each column
print(df_data.dtypes)
```

```python
# numerical columns vs categorical columns

numerical_columns = [col for col in df_data.columns if df_data[col].dtype != 'object']
categorical_columns = [col for col in df_data.columns if df_data[col].dtype == 'object']
```

# EDA 
1.univarite analysis

2.bivariate analysis

3.multivariate analysis


```python
# first check the numerical columns check the distribution use histogroms

for col in numerical_columns:
    sns.histplot(df_data[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# catagorical columns check the distribution use countplot

for col in categorical_columns:
    sns.countplot(x=df_data[col])
    plt.figure(figsize=(15, 8))
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()
```
